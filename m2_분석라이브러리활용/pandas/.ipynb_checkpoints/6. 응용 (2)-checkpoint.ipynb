````python
def add_10(x):
    return x + 10

def add_two_obj(x, y):
    return x + y

print(add_10(10))
print(add_two_obj(10, 10))
````

````python
def mul_10(x):
    return x * 10

def div_two_obj(x, y):
    return x / y

print(mul_10(20))
print(div_two_obj(200,20))
````

### Q9. titanic 데이터셋을 seaborn에서 로딩한 후 블린 인덱싱을 이용하여 아래 사항들을 수행하세요.
- 나이가 10대(10~19세)인 승객만 따로 선택
- 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택
- 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택


````python
import pandas as pd
import seaborn as sns 

titanic = sns.load_dataset('titanic')
titanic.head()
````

````python
#나이가 10대(10~19세)인 승객만 따로 선택
titanic[(titanic['age'] >= 10) & (titanic['age'] <= 19)]

#mask1 = (titanic['age'] >= 10) & (titanic['age'] <= 19)
#titanic.loc[mask1 :]
````

````python
#나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택
titanic[(titanic['age'] < 10) & (titanic['sex'] == 'female')] 
````

````python
#나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택
df_t = titanic[(titanic['age'] < 10) | (titanic['age'] >= 60)] 
df_t[['age', 'sex', 'alone']]

#df_t = titanic.loc[(titanic['age'] < 10) | (titanic['age'] >= 60), [['age', 'sex', 'alone']]   
````

#### 그룹연산
- 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며, 이런 방식으로 분할 처리 하는 것을 그룹 연산이라함.
- 그룹 연산은 데이터를 집계, 반환, 필터하는 데 효율적이며, 판다스 groupby() 메서드를 사용
- 그룹 객체 만들기(분할)
- 그룹 연산 메서드(적용-결합)


````python
import pandas as pd
import seaborn as sns 

titanic = sns.load_dataset('titanic')
titanic.columns
````

````python
titanic.head() #pclass랑 class는 같은거임 이렇게 중복되는건 없애주는게 좋음
#새로운 정보력도 없으면서 용량만 차지하는건 모델을 더 악화시킴 
````

````python
df = titanic.loc[:,['age', 'sex', 'class', 'fare', 'survived']]

print('승객 수:', len(df))
df.head()
````

````python
df.columns
````

````python
#class열을 기준으로 그룹핑(분할)
grouped = df.groupby('class')
print(grouped)
list(grouped)
````

````python
#그룹 객체 iteration(반복 객체)으로 출력
for key, group in grouped:
    print('* key :', key)
    print('* number :', len(group)) #그룹의 수 
    print(group.head(), '\n')
````

````python
#연산 메서드
average = grouped.mean()
 #survived는 0과 1로 구성되어있는데 얘로 평균이 나오니까 생존률로 봐도 됨 
average
````

````python
stat = grouped.max() #알파벳도 순서대로 카운트해서 더 큰값이 나온것 m이 더 큰가벼
stat
````

````python
#Q.3rd 그룹만을 선택해서 group3으로 이름으로 저장하고 통계요약표를 출력
group3 = grouped.get_group('Third') #get_group 함수 -> 그룹by한거 안에서 특정 그룹만을 꺼냄 
group3.describe()

#fare는 요금임 
````

````python
grouped_two = df.groupby(['class', 'sex']) # 두 그룹으로 그룹바이
#class는 3분류고, sex는 2분류니까 총 3*2 = 6분류겠네
#얘도 for문 가능하겠지

average_two = grouped_two.mean()
average_two
````

````python
group3f = grouped_two.get_group(('Third', 'female')) #2차원이니까 괄호 2개
group3f.head()
````

````python
#모든 열에 여러 함수를 매핑
agg_all = grouped.agg(['min', 'max'])
agg_all.head()
````

````python
agg_sep = grouped.agg({'fare':['min', 'max'], 'age':'mean'})
agg_sep.head()
````

    #Q class로 그룹핑한 grouped에서 데이터 개수가 200개 이상인 그룹만을 필터링하여 데이터 프레임으로 반환하여grouped_filter 이름으로 저장후 처음 5개행을 출력

## pivot
피벗 테이블을 구성하는 4가지 요소(행인덱스, 열인덱스, 데이터 값, 데이터 집계함수)에 적용할 데이터 프레임의 열을 각각 지정하여 함수의 인자로 전달

````python
#pivot 

import pandas as pd
import seaborn as sns 

titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age', 'sex', 'class', 'fare', 'survived']]
df.head()
````

````python
pdf1 = pd.pivot_table(df, index = 'class', columns = 'sex', values = 'survived', aggfunc = 'mean')  
#aggfunc -> values에 무슨 계산을 해줄건데 
pdf1.head()
````

````python
pdf1 = pd.pivot_table(df, index = 'class', columns = 'sex', values = 'age', aggfunc = 'mean')  
#aggfunc -> values에 무슨 계산을 해줄건데 
pdf1.head()
````

````python
pdf1 = pd.pivot_table(df, index = 'class', columns = 'sex', values = 'fare', aggfunc = 'mean')  
#aggfunc -> values에 무슨 계산을 해줄건데 
pdf1.head()
````

````python
#행, 열, 값에 사용할 열을 2개 지정 - index, values, aggfunc에 각각 2개 이상 지정

pdf2 = pd.pivot_table(df, index = ['class', 'sex'], columns = 'survived', values = ['age', 'fare'], aggfunc = ['mean', 'max'])
pdf2.head()

#만약 pdf2에서 first에서 female age는 1등석 여자 중 죽은사람중 평균 나이 이런거 
````

## [pipe vs applymap vs apply]  

pipe와 applymap과의 차이는 반환하는 결과가 pipe가 더 다양하다는 것으로 applymap은 데이터프레임 원소에 적용하므로 데이터프레임으로만 리턴되지만 pipe는 사용자 함수의 리턴값에 따라 개별 값, 시리즈, 데이터프레임을 반환할 수 있다.  

현재 가장 많이 쓰이는 메서드는 apply인데 그 이유는 다양한 형태로 값을 입력 받을 수 있고, 반환하는 형태도 폭이 넓기 때문. 결과는 같아보이지만, 시리즈를 이용해 데이터프레임의 행과 열도 이용할 수 있으며 시리즈 그 자체도 함수로 활용할 수 있다. 

만약 데이터프레임에 함수를 적용해야한다면 리턴하는 형태가 많은 pipe 사용도 고려


<img src = 'dataset/apply_applymap_pipe.jpg' STYLE = 'width:800px;'>

````python
import pandas as pd
df = pd.DataFrame({'Col1':[30,40,50,60],'Col2':[23,35,65,45],'Col3':[85,87,90,89]},index = list('ABCD'))
df
````

````python
df.apply(lambda x: x.sum(), axis = 0 ) #시리즈가 리턴됨
````

````python
df.Col1.apply(lambda x:x+1)
````

````python
#시리즈 타입의 경우 map 함수도 적용 가능했어
df.Col1.map(lambda x : x+1)
````

````python
df.applymap(lambda x:'%.2f' %x)
````

````python
# The pipe() method allows you to apply one or more functions to the DataFrame object.
# 데이터 프레임 객체에 하나 이상의 기능을 적용할 때 사용한다 
import pandas as pd

def change_age(x):
    x["age"]=[10, 20, 30]
    return x

data = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

df = pd.DataFrame(data)
df
````

````python
df.pipe(change_age)
````

````python
from sklearn.datasets import load_boston
import warnings
warnings.filterwarnings('ignore')

boston = load_boston()
print(boston.keys(), '\n')    #target - 주택 가격 
print(boston.feature_names) #독립변수 칼럼 - 주택 가격에 영향을 미치는 요인들
df_boston = pd.DataFrame(boston.data, columns = boston.feature_names)
df_boston['target'] = boston.target
df_boston.head()
````

````python
from sklearn.preprocessing import MinMaxScaler #MinMaxScaler 얘는 클래스니까 객체를 만들어줘야지

scaler = MinMaxScaler()  # max = 1, min = 0으로 만들어줌  
df_boston['RM'] = scaler.fit_transform(df_boston[['RM']])  #fit - 스케일링 변환을 할 준비를 하는것 , transform은 그걸 실행하는것
df_boston[['RM']].describe()
````

# encoding 
- 문자로 되어있는 것을 숫자로 바꿔준다 
- LableEncoder, OneHotEncoder

````python
import pandas as pd
import numpy as np
import seaborn as sns

titanic = sns.load_dataset('titanic')
#특정 컬럼의 값을 변경
titanic['sex'] = titanic.sex.replace({'male':0, 'female':1})
#열이름 변경
titanic.rename(columns = {'sex': 'gender', 'age':'age_cat', 'class':'c_level'}, inplace = True)
#컬럼명을 대문자로 변경
titanic.columns = map(lambda x: str(x).upper(), titanic.columns)
#컬럼명을 소문자로 변경
titanic.rename(columns = str.lower, inplace = True)
df = titanic[['survived', 'pclass', 'gender', 'c_level']]
df.head()
````

````python
titanic.columns
````

````python
df1 = df.copy()
````

````python
import warnings
warnings.filterwarnings('ignore')
df['gender'].replace([1,0],['female', 'male'], inplace = True)
df.head()
````

````python
#인코딩 - LableEncoder
from sklearn.preprocessing import LabelEncoder

#LabelEncoder  얘는 클래스니까 객체 만들어줘 
le = LabelEncoder()
features = ['gender', 'c_level'] #얘네를 숫자로 만들어야지 ~! , for문써서 한번에 돌리려고 이렇게 리스트 만들어준것

for f in features:
    df1[f] = le.fit_transform(df[f]) 
df1.head()
````

````python
#알아서 범주가 2개면 0,1 / 범주가 3개면 0,1,2로 만들어줌 LabelEncoder 얘가
#확인해볼까
df1.c_level.value_counts()
````

````python
#만약에 범주가 99개야 이걸 LabelEncoder 얘로 해버리면 숫자가 뭐 99개니까 숫자로 분류 시켜준게 의미가 없잖아 
#이렇게 되면 범주로서의 숫자 의미가 아니라 크기로서의 숫자로 들어가서 알고리즘에 영향을 준다고 

#이럴때는 OneHotEncoder
#얘는 gender -1 , gender - 2 로 칼럼으로 올라가고 , 해당되지않는건 0, 해당 되는건 1 로 해서 함 
#삘이 그거 r에서 배운 더미변수? 그거랑 비슷한 느낌인거같음 
#맞네 더미네 
````
