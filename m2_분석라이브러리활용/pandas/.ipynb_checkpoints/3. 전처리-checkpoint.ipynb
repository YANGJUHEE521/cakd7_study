```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
df.age.value_counts()
```

```python
#범주화 
def get_category(x):
    cat = ''
    if x <20:
        cat = 'child'
    elif x < 30:
        cat = 'Young Adult'
    elif x < 50:
        cat = 'Adult'
    else:
        cat = 'Elderly'
    return cat
df['age_class'] = df.age.apply(lambda x:get_category(x))
df[['age', 'age_class']].head()
```

```python
#sex 칼럼을 대문자로 변환 
def big(x):
    return x.upper()

df['sex'] = df.sex.apply(lambda x:big(x))
df[['sex']].head()
```
