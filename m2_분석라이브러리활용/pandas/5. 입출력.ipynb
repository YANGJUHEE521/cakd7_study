```python
df = pd.read_csv('dataset/test.csv', index_col = 0) #여기서 인덱스 없애고 싶으면 index_col = 0
df
```

```python
#json #딕셔너리 형태로 되어있음 
df1 = df.copy()
df1.to_json('dataset/test1.json')
df1_json = pd.read_json('dataset/test1.json')
df1_json
```
