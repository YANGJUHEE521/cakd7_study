## 판다스
- 판다스는 구조화된 데이터 형식을 제공
- 시리즈는 1차원 배열, 데이터 프레임은 2차원 배열
- 라이브러리는 여러 종류의 class와 다양한 내장 함수로 구성
- 시리즈와 데이터프레임은 대표적인 클래스 객체임 
- 시리즈 인덱스는 데이터 값과 일대일 대응, 파이썬 딕셔너리와 비슷한 구조 
- contents
    - Series(변환, 인덱스 구조, 원소 선택)
    - DataFrame(변환,행 인덱스/열 이름 지정, 삭제, 선택, 추가, 변경, 전치, 인덱스 활용)

cf) 클래스는 붕어빵 기계, 붕어빵은 객체, 
클래스를 사용할 땐, 클래스를 객체로 만들어서 객체를 통해서 객체한테 클래스 자원을 쓸 수 있는 권한 부여



- 시리즈는 하나의 벡터로 되어있음 
- 스칼라는 하나의 값이고 이게 모여서 하나의 벡터가 된다고 생각
- 이 벡터들이 여러개 있는 행렬(2차원 matrix)의 형태가 데이터 프레임

````python
#딕셔너리 -> 시리즈
import pandas as pd

dict_data = {'a': 1, 'b': 2, 'c': 3}
sr = pd.Series(dict_data)
print(sr)
print(type(sr))
````

````python
#리스트 -> 시리즈 

list_data = ['2019-07-02', 3.14, 'ABC', 100, True]
sr = pd.Series(list_data, index = list('abcde')) #한꺼번에 이렇게 붙여서 쓸 수 있음
sr                        #index 안 주면 기본으로 0,1,2 .. 들어감 
````

````python
#인덱스 및 값 

idx = sr.index
val = sr.values
print(idx)
print(val)
````

````python
#튜플을 시리즈로 변환
tup_data = ('kevin', '2019-07-02', '남', True)
sr = pd.Series(tup_data, index = ['이름', '생년월일', '성별', '학생여부'])
print(sr)
````

````python
print(sr[0])
print(sr['이름'])
print(sr[[1, 2]]) #리스트로 묶어서 인덱싱 가능
print(sr[1,2]) #그냥 이렇게 쓰면 에러뜸 
````

## 데이터 프레임
- 2차원 배열, R의 데이터 프레임에서 유래 
- 열은 각각 시리즈 객체
- 시리즈를 열벡터라고 하면 데이터 프레임은 여러 개의 열 벡터들이 같은 행 인덱스를 기준으로 줄지어 결합된 2차원 벡터 또는 행렬 
- 선형대수학에서 열 벡터(m x 1 행렬)는 m 원소들의 단일 열 행렬
- 행 벡터(1 x m 행렬)은 m 원소들의 단일 행 행렬
- 리스트, 딕셔너리, ndarray 등 다양한 데이터로부터 생성
- 반대로 리스트, 딕셔너리, ndarray 등으로 변환될 수 있음 

````python
#배열을 데이터프레임으로 변환
np.random.seed(0) #난수를 해줘서 랜덤 값이 안 바뀌게
data = np.random.randint(100, 120, size = (3,3)) #100~120 사이에 3x3 정수 행렬 
print(data, type(data)) #데이터 타입은 numpy 배열 

df = pd.DataFrame(data, index = ['d1', 'd2', 'd3'], columns = ['pd', 'sales', 'inv'])
df #열은 3x1 벡터, 행은 1x3 벡터 
````

````python
#인덱싱 : iloc 정수 인덱스 , loc 이름 인덱스 
print(df.iloc[1], '\n')
print(df.loc['d2'])
````

````python
# Q. iloc, loc를 사용하여 107을 출력하세요.
print(df.iloc[1,2], df.loc['d2', 'inv'])
````

````python
df1 = df.copy() #복사본이 만들어짐  
df1
````

````python
#d3의 값들을 모두 0으로 만드세요. 
df1.loc['d3'] = 0
df1
````

````python
# 값들이 모두 0인 d4 행을 추가하세요. 
df1.loc['d4'] = 0
df1
````

````python
#1행 1열의 값을 0으로 만드세요.
df1.iloc[1,1] = 0
df1
````

````python
import pandas as pd
import numpy as np
````

````python
#id, gender, age, region
id = np.arange(1, 1001) #1~ 1000
i1 = pd.Series(id)
gender = np.random.randint(2, size = 1000) #남 0 , 여 1
g1 = pd.Series(gender)
age = np.random.randint(1, 101, size = 1000) #1~100살 까지 
a1 = pd.Series(age)
region = np.random.randint(1,11,size = 1000) #1~10번 지역을 1000개 
r1 = pd.Series(region)

df1 = pd.concat([i1, g1, a1, r1], axis = 1) #axis = 1 열 방향 
df1.rename(columns = {0: 'id', 1: 'gender', 2: 'age', 3: 'region'}, inplace = True) #열 이름을 딕셔너리로 지정{열번호: 열이름},  inplace = True -> 원본에 반영하라는 뜻 
df1.head()
````

### 1 공복혈당

````python

blood_sugar = np.random.randint(70, 250, size = 1000) 
b1 = pd.Series(blood_sugar)
````

### 2 혈청크레아티닌 - 신장기능 지표
```
정상일 때)
남성 : 0.61~1.04mg/dL
여성 : 0.47~0.79mg/dL

cf) 혈중 크레아티닌 농도의 정상범위는 0.50~1.4 mg/dL 입니다.
코드 cf) https://abluesnake.tistory.com/114
```

````python
from numpy import random
SCr = (2) * random.random_sample(size=1000)
s1 = pd.Series(SCr)
````

### 3 bmi(비만도)
```
8.5 ~ 22.9: 정상
23.0 ~ 24.9 : 과체중
25.0 ~ 29.9 : 비만 
30 이상 : 고도비만
```

````python
bmi = (35 - 8) * random.random_sample(size=1000) + 8
m1 = pd.Series(bmi)
````
