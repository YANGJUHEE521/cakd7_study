```python
import numpy as np
import pandas as pd
```

```python
#고객별 속성 데이터

# cusno : 고객번호 1000개 랜덤 
# gender : 0,1 정수 난수 생성
# age : 0 ~ 80사이 정수 난수 생성
# region : 1 ~ 10 사이 정수 난수 생성
np.random.seed(0)
df = pd.DataFrame(list(range(1, 1001)), columns = ['id'])
df['cusno'] = np.random.randint(1000, 10000, 1000) #4자리수로 설정
df['gender'] = np.random.choice([0, 1], 1000)
df['age'] = np.random.randint(0, 81, 1000)
df['region'] = np.random.randint(1, 11, 1000)
df
```

```python
import random
# 거래 데이터

# product_20 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10,20,30
# product_21 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10,20,30
# price_20 : 1000 ~ 50000 사이 실수 난수 생성
# price_21 : 1000 ~ 50000 사이 실수 난수 생성
# quantity_20 : 1 ~ 100 사이 정수 난수 생성
# quantity_21 : 1 ~ 100 사이 정수 난수 생성
# amount_20 : price_20 * quantity_20
# amount_21 : price_21 * quantity_21
# sales : 20년 대비 21년구매금액이 증가면 1, 감소면 0
np.random.seed(1)
df['product_20'] = np.random.choice([10, 20, 30], 1000)
df['product_21'] = np.random.choice([10, 20, 30], 1000)
r = random.random()
df['price_20'] = (np.random.randint(1000, 50001, 1000) * r).round(1)
df['price_21'] = (np.random.randint(1000, 50001, 1000) * r).round(1)
df['quantity_20'] = np.random.randint(1, 101, 1000)
df['quantity_21'] = np.random.randint(1, 101, 1000)
df['amount_20'] = df['price_20'] * df['quantity_20']
df['amount_21'] = df['price_21'] * df['quantity_21']

df['sales'] = np.where(df['amount_20'] < df['amount_21'], 1, 0)
df
```

```python
#먼저, 구분제품코드랑 제품 가격을 각각 생성해줘서 제품 가격에 따라 구분제품코드를 바꿔야함 



def ran(x):
    if x < 15000:
        return 10
    elif x < 30000:
        return 20
    else:
        return 30
  
df['product_20'] = df.price_20.apply(lambda x : ran(x))
df['product_21'] = df.price_21.apply(lambda x : ran(x))
df['product_20'].value_counts()
```

```python
df['product_21'].value_counts()
```

```python
df['sales'].value_counts()
```

```python
# 컬럼 추가 
#- 구매 제품 가격대가 더 비싸졌으면 1, 아니면 0 
# - 제품 수량이 더 늘었으면 1, 아니면 0 
# 구매 가격도 늘고 제품 수량도 더 많이 산 사람 - Gold, 아니면 General - 편하게 멤버십이라고 칭하겠음


df['product_flow'] = np.where(df['product_20'] < df['product_21'], 1, 0)
df['quantity_flow'] = np.where(df['quantity_20'] < df['quantity_21'], 1, 0)
df['membership'] = np.where((df['product_flow'] == 1) & (df['quantity_flow'] == 1), 'Gold', 'General')
df.head()
```

```python
#나이대 칼럼 추가
def get_category(x):

    if x <10:
        return 'child'
    elif x < 20:
        return 'teenager'
    elif x < 30:
        return 'Young Adult'
    elif x < 40:
        return 'Adult'
    else:
        return 'Elderly'
 
df['age_class'] = df.age.apply(lambda x:get_category(x))
df.head()
```
