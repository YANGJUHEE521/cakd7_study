# 나이 전처리

## 호칭 전처리
```
Mr. - 혼인 여부 관계 없이 남성
Ms. - 혼인여부 관계없이 여성 + Mlle
Mrs. - 혼인을 한 여성 + Mme
Miss - 미혼인 여성  
```

````python
#name에서 성별 호칭 칼럼 생성 
import re 


idx = 0

for t in titanic_df['name']:
    m = re.search('[,]\s+([\w]+)', t)
    titanic_df.loc[idx, 'title'] = m.group(1)
    idx += 1

titanic_df['title'].unique()
````

````python
titanic_df['title'].value_counts()
````

# 강사님이랑 예측

````python
#분석용 data set 
titanic_df = pd.read_pickle('tdf.pkl')
titanic_df.head() #OneHotEncoding 했구만 
````

````python
from sklearn.model_selection import train_test_split  #train_test_split 훈련용 데이터와 테스트용 데이터를 나눈다

# 독립변수, 종속변수 분리
y_tdf = titanic_df['survived'] #종속변수 
X_tdf = titanic_df.drop('survived', axis = 1) #독립변수 

#학습용 data와 평가용 data를 8:2로 분리( test_size = 0.2)
X_train, X_test, y_train, y_test = train_test_split(X_tdf, y_tdf, test_size = 0.2, random_state =11) #random_state =11 - 샘플 고정 
print(X_train.shape, X_test.shape) #8:2로 잘 나누어 졌고, (,8) 뒤에 8은 칼럼 수 
````

````python
#모델 학습 및 평가 
from sklearn.metrics import accuracy_score #분류 모델의 가장 기본적인 평가 방법
from sklearn.linear_model import LogisticRegression  # linear_model - 선형 회귀 모델 ,  LogisticRegression- 선형 회귀모델을 이용해서 분류 모델
from sklearn.ensemble import RandomForestClassifier # RandomForestClassifier - 의사결정나무를 축소해서 여러개의 의사결정나무를 만들어서 조금 더 치우치지 않게끔 알고리즘 수정해서 만든것 

rf_model = RandomForestClassifier() #객체 생성 #알고리즘에 있는 모든 자원을 쓸 수 있지 
#학습 - fit, 학습 하려면 데이터를 집어넣어야하니까 인자로 학습용 데이터를 집어넣어 
rf_model.fit(X_train, y_train) #주어진 애들로 학습을 해, 분류할 수 있게 됨 -> 예측할 수 있게됨 
rf_pred = rf_model.predict(X_test) #y 값을 예측하려면 일단 평가를 해야하니까 독립변수의 평가용 데이터를 인자로 넣자 
#정확도 평가 - 실제값, 예측값을 인자로 넣어줘 
accuracy_rf = accuracy_score(y_test, rf_pred).round(2)


lr_model = LogisticRegression() #객체 생성 #알고리즘에 있는 모든 자원을 쓸 수 있지 
#학습 - fit, 학습 하려면 데이터를 집어넣어야하니까 인자로 학습용 데이터를 집어넣어 
lr_model.fit(X_train, y_train) #주어진 애들로 학습을 해, 분류할 수 있게 됨 -> 예측할 수 있게됨 
lr_pred = lr_model.predict(X_test) #y 값을 예측하려면 일단 평가를 해야하니까 독립변수의 평가용 데이터를 인자로 넣자 
#정확도 평가 - 실제값, 예측값을 인자로 넣어줘 
accuracy_lr = accuracy_score(y_test, lr_pred).round(2)

print(f'rf 정확도: {accuracy_rf}, lr 정확도:{accuracy_lr}')
````
